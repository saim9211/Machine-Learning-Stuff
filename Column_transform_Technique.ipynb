```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("covid_toy.csv")
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('has_covid',axis=1),df['has_covid'],test_size=0.3)
X_train.shape,X_test.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer
ct=ColumnTransformer(
    transformers=[
        ('tf1',SimpleImputer(),['fever']),
        ('tf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
        ('tf3',OneHotEncoder(sparse_output=False),['gender','city'])
    ],
    remainder='passthrough'
)
ct.fit_transform(X_train)
ct.fit_transform(X_test)
```

```python
t=ct.set_output(transform="pandas")
t
```
