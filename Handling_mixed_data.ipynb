```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.sample(3)
```

```python
df['number_int'] = pd.to_numeric(df['number'].str.extract('(\\d+)')[0], errors='coerce')
```

```python
df['number_Str'] = df['number'].str.extract(r'([^\d]+)')[0]
```

```python
df['number_int'].fillna(0, inplace=True)
```

```python
df['Cabin_int'] = pd.to_numeric(df['Cabin'].str.extract('(\\d+)')[0], errors='coerce')
df['Cabin_Str'] = df['Cabin'].str.extract(r'([A-Za-z]+)')[0]
```

```python
df.head()
```

```python
df['Ticket_int'] = pd.to_numeric(df['Ticket'].str.extract('(\\d+)')[0], errors='coerce')
df['Ticket_Str'] = df['Ticket'].str.extract(r'([^\\d]+)')[0]
```
