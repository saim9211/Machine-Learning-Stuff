```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
x_train.value_counts()
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
print("mean absolute error",mean_absolute_error(y_test,y_pred))
print("mean squared error",mean_squared_error(y_test,y_pred))
```

```python
print("r2_Score",r2_score(y_test,y_pred))
```

```python
r2 = r2_score(y_test, y_pred)
n = len(y_test)
p = x_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))
print("Adjusted R2 Score:", adjusted_r2)
```

```python
df['random_feature'] = np.random.rand(len(df))

# Redefine features (X) to include the new random column
X = df[['cgpa', 'random_feature']]
y = df['package']

# Split the data into training and testing sets
x_train_new, x_test_new, y_train_new, y_test_new = train_test_split(X, y, test_size=0.2, random_state=2)

# Train a new Linear Regression model
lr_new = LinearRegression()
lr_new.fit(x_train_new, y_train_new)

# Make predictions with the new model
y_pred_new = lr_new.predict(x_test_new)

# Calculate R2 score for the new model
r2_new = r2_score(y_test_new, y_pred_new)
print(f"R2 Score with random feature: {r2_new}")

# Calculate Adjusted R2 score for the new model
n_new = len(y_test_new)
p_new = x_test_new.shape[1] # Number of predictors (cgpa + random_feature)
adjusted_r2_new = 1 - ((1 - r2_new) * (n_new - 1) / (n_new - p_new - 1))
print(f"Adjusted R2 Score with random feature: {adjusted_r2_new}")
```

```python
df['cgpa_squared'] = df['cgpa']**2

# Redefine features (X) to include 'cgpa' and the new 'cgpa_squared' column
X_poly = df[['cgpa', 'cgpa_squared']]
y_poly = df['package']

# Split the data into training and testing sets for the polynomial model
x_train_poly, x_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y_poly, test_size=0.2, random_state=2)

# Train a new Linear Regression model with the polynomial feature
lr_poly = LinearRegression()
lr_poly.fit(x_train_poly, y_train_poly)

# Make predictions with the new model
y_pred_poly = lr_poly.predict(x_test_poly)

# Calculate R2 score for the new model
r2_poly = r2_score(y_test_poly, y_pred_poly)
print(f"R2 Score with cgpa_squared feature: {r2_poly}")

# Calculate Adjusted R2 score for the new model
n_poly = len(y_test_poly)
p_poly = x_test_poly.shape[1] # Number of predictors (cgpa + cgpa_squared)
adjusted_r2_poly = 1 - ((1 - r2_poly) * (n_poly - 1) / (n_poly - p_poly - 1))
print(f"Adjusted R2 Score with cgpa_squared feature: {adjusted_r2_poly}")
```

```python
df = pd.read_csv("/content/AmesHousing.csv")
# Filtering the DataFrame for the 'Big 6' high-signal features
df_useful = df[['Overall Qual', 'Gr Liv Area', 'Garage Cars', 'Total Bsmt SF', 'Full Bath', 'Year Built']].copy()
```

```python
# Impute missing values in df_useful before splitting
for col in ['Garage Cars', 'Total Bsmt SF']:
    df_useful[col] = df_useful[col].fillna(df_useful[col].median())

x_train,x_test,y_train,y_test=train_test_split(df_useful,df['SalePrice'],test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2:.4f}")

n = len(y_test)
p = x_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))
print(f"Adjusted R-squared: {adjusted_r2:.4f}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', lw=2, label='Perfect Prediction')
plt.xlabel('Actual SalePrice')
plt.ylabel('Predicted SalePrice')
plt.title('Actual vs. Predicted SalePrice')
plt.legend()
plt.grid(True)
plt.show()
```
