### for converting the values 

```python
def rename(name):
    try:
        if int(name) == 2024:
            return 'leap'
        else:
            return name
    except:
        return name  # in case of empty or invalid data

df = pd.read_csv("New1.csv", converters={'Year': rename})
print(df)
```
